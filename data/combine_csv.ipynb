### Data Cleaning Script
Cleaning up the data to load into Unity; it's uh kinda messy, but it works.

```python
import pandas as pd

# ignores special country cases in main dataset
tu = pd.read_csv('TIME_USE_OECD_EXTENDED.csv')
tu['Country'] = tu['Country'].apply(lambda x: x[:-1] if '*' in x else x)
tu.head(5)
```

```python
def is_in_col(string, row, columns):
  return any(string in str(row[col]) for col in columns)

def sum_columns(df, columns):
  return df.apply(lambda row: sum(int(row[col]) for col in columns) if not is_in_col('..', row, columns) else '..', axis=1)

# read in the secondary dataset
bli = pd.read_csv('BETTER_LIFE_INDEX_OECD.csv')
bli.head(3)

# join the two datasets
merged = pd.merge(tu, bli, on='Country', how='inner')
merged['Other Paid Work'] = sum_columns(merged, ['research/homework', 'job search', 'other paid work or study-related'])
merged['Care for Others'] = sum_columns(merged, ['care for household members', 'care for non household members', 'volunteering'])
merged['Housework'] = sum_columns(merged, ['travel related to household activities', 'routine housework'])
merged['Events'] = sum_columns(merged, ['shopping', 'other unpaid'])
merged['Unpaid Work'] = sum_columns(merged, ['sports', 'participating / attending events'])
merged.head(3)
```
